```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
import numpy as np
import time

# Load the dataset
df = pd.read_csv('creditcard.csv')

# Feature scaling: Scaling 'Amount' and 'Time'
scaler = StandardScaler()
df['scaled_amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['scaled_time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))

# Drop the original columns after scaling
df = df.drop(['Amount', 'Time'], axis=1)

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the enhanced correlation matrix with annotations
plt.figure(figsize=(15,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix with Annotations')
plt.show()

# Define a threshold for selecting significant variables (correlation > 0.1)
significant_corr_threshold = 0.1
correlation_with_target = correlation_matrix['Class'].drop('Class')

# Identify significant features based on the threshold
significant_features = correlation_with_target[abs(correlation_with_target) > significant_corr_threshold].index.tolist()
insignificant_features = correlation_with_target[abs(correlation_with_target) <= significant_corr_threshold].index.tolist()

print("Significant Features (Correlation > 0.1):\n", significant_features)
print("\nRemoved Insignificant Features (Correlation <= 0.1):\n", insignificant_features)

# Keep only significant features
df_significant = df[significant_features + ['Class']]

# Define X (features) and y (target)
X = df_significant.drop('Class', axis=1)
y = df_significant['Class']

# Handle the imbalanced dataset using SMOTE
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Measure training time
start_train_time = time.time()

# Train a Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

end_train_time = time.time()

# Measure prediction time for the test set
start_test_predict_time = time.time()

# Make predictions on the test set
y_pred = model.predict(X_test)

end_test_predict_time = time.time()

# Evaluate the model performance
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Calculate and print model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Print training and prediction times
print(f"\nTraining Time: {end_train_time - start_train_time:.4f} seconds")
print(f"Prediction Time (for test set): {end_test_predict_time - start_test_predict_time:.4f} seconds")

# Now, let's predict for a new transaction

# Example new transaction data (replace with actual values)
new_data = pd.DataFrame({
    'Time': [10000],  # Replace with actual transaction time
    'V1': [-1.359807],  'V2': [-0.072781],  'V3': [2.536346], 'V4': [1.378155], 'V5': [-0.338321],
    'V6': [0.462388],  'V7': [0.239599],  'V8': [0.098698],  'V9': [0.363787],  'V10': [0.090794],
    'V11': [-0.5516],  'V12': [0.415798],  'V13': [-0.187257],  'V14': [0.069083],  'V15': [0.101288],
    'V16': [0.237609],  'V17': [-0.377869],  'V18': [0.505056],  'V19': [0.240543],  'V20': [0.252325],
    'V21': [-0.194401],  'V22': [0.261057],  'V23': [-0.143645],  'V24': [-0.206251],  'V25': [-0.190321],
    'V26': [0.159089],  'V27': [0.129697],  'V28': [-0.083082],
    'Amount': [123.45]  # Replace with actual transaction amount
})

# Feature scaling: Apply the same scaling as the training data
new_data['scaled_amount'] = scaler.transform(new_data['Amount'].values.reshape(-1, 1))
new_data['scaled_time'] = scaler.transform(new_data['Time'].values.reshape(-1, 1))

# Drop the original 'Amount' and 'Time' columns after scaling
new_data = new_data.drop(['Amount', 'Time'], axis=1)

# Select only the significant features for prediction
new_data_significant = new_data[significant_features]

# Measure prediction time for the new transaction
start_new_predict_time = time.time()

# Predict whether the new transaction is fraudulent
prediction = model.predict(new_data_significant)

end_new_predict_time = time.time()

# Interpret the prediction
if prediction[0] == 1:
    print("The transaction is fraudulent.")
else:
    print("The transaction is not fraudulent.")

# Print prediction time for the new transaction
print(f"Prediction Time (for new transaction): {end_new_predict_time - start_new_predict_time:.4f} seconds")
```
